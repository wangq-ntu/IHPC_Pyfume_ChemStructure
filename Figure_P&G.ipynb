```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample DataFrame
data = {
    'categories': ['A', 'B', 'C', 'D'],
    'values1': [20, 35, 30, 25],
    'values2': [15, 25, 20, 30]
}
df = pd.DataFrame(data)

# Define color maps
color_map1 = plt.cm.viridis
color_map2 = plt.cm.plasma

# Plotting multiple columns
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
bar_width = 0.35
indices = np.arange(len(df['categories']))
plt.bar(indices - bar_width/2, df['values1'], bar_width, color=color_map1(0.6), label='Data 1')
plt.bar(indices + bar_width/2, df['values2'], bar_width, color=color_map2(0.6), label='Data 2')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('2D Column Chart')
plt.xticks(indices, df['categories'])
plt.legend()  # Show legend for different columns

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

import numpy as np

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]

# Set position of bar on X axis
br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, IT, color ='r', width = barWidth,
        edgecolor ='grey', label ='IT')
plt.bar(br2, ECE, color ='g', width = barWidth,
        edgecolor ='grey', label ='ECE')
plt.bar(br3, CSE, color ='b', width = barWidth,
        edgecolor ='grey', label ='CSE')

# Adding Xticks
plt.xlabel('Branch', fontweight ='bold', fontsize = 15)
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(IT))],
        ['2015', '2016', '2017', '2018', '2019'])

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Data 1': [20, 35, 30, 35, 27],
    'Data 2': [25, 32, 34, 20, 25]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the data
df.set_index('Category').plot(kind='bar', ax=ax)

# Adjust the spacing between subplots
plt.tight_layout()

# Add labels and title
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Multiple Bar Plots')

# Rotate the x-axis labels for better visibility
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
